```python
def stints_to_dummy_simple(data, df_on_court):
    
    data_stints = pd.merge(data, df_on_court)
    
    # seperate the scores
    scores = data_stints[['GAME_ID', 'stint', 'HOME_PTS', 'AWAY_PTS', 'HOME_PM', 'HOME_PM_100']]
    
    # home and away names
    home_names = ['GUARD_1_HOME', 'GUARD_2_HOME', 'FORWARD_1_HOME', 'FORWARD_2_HOME', 'CENTER_1_HOME']
    away_names = ['GUARD_1_AWAY', 'GUARD_2_AWAY', 'FORWARD_1_AWAY', 'FORWARD_2_AWAY', 'CENTER_1_AWAY']
    all_names = ['GUARD_1_HOME', 'GUARD_2_HOME', 'FORWARD_1_HOME', 'FORWARD_2_HOME', 'CENTER_1_HOME', 'GUARD_1_AWAY', 'GUARD_2_AWAY', 'FORWARD_1_AWAY', 'FORWARD_2_AWAY', 'CENTER_1_AWAY']

    # shift into long format, id: game_id & stint - value vars home and away
    tmp = pd.melt(data_stints, id_vars=['GAME_ID', 'stint'], value_vars=all_names)
    
    # get the dummies on the long format
    tmp2 = pd.get_dummies(tmp, columns=['value'], prefix="", prefix_sep="")
    
    # create boolean and factor vector
    bool_away = tmp2['variable'].isin(away_names)
    weight_vec = np.where(bool_away, -1, 1).reshape(-1, 1)
    
    # extract dummy matrix
    tmp_a = np.asarray(tmp2.iloc[:, 3:])
    
    # adjust for home and away
    tmp_b = tmp_a * weight_vec
    
    # replace
    tmp2.iloc[:, 3:] = tmp_b
    
    # merge scores and players
    df = pd.merge(scores, tmp2)
    
    # get series objects to track the players and their "position"
    series_location_player = pd.Series(df.columns[7:].values)
    series_id_player = series_location_player.apply(lambda x: x.replace('.0', ''))
    series_id_player = series_id_player.astype(int)

    return df, series_id_player
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
schedule = pd.read_csv("./data/season_prediction/schedule.csv", dtype={'GAME_ID':str})

from nba_api.stats.endpoints.gamerotation import GameRotation
game_id = "0022001079" #"0021800001"
call = GameRotation(game_id)

rotation = pd.concat(call.get_data_frames())
```

```python
def generate_dummy(rotation, schedule, game_id):
    
    """ insert subseted rotation, schedule and the corresponding game id
    from rotation and schedule data arrive at data, starters, starting_lineup and subs
    """

    # extract home games
    home_game = schedule[['TEAM_ID', 'GAME_ID', 'is_home_game']]

    # extract in and out times
    out_times = np.sort(rotation['OUT_TIME_REAL'].unique()) # get all out times
    in_times = np.sort(rotation['OUT_TIME_REAL'].unique()) # get all in times

    assert np.all(out_times == in_times), "Out and In times do not correspond"

    # sub time and time post
    sub_time = out_times
    time_post = np.arange(1, len(sub_time) + 1)

    # helper
    tmp_tp = pd.DataFrame(data={'sub_time':sub_time, 'time_post':time_post})

    # add sub times and home game, create player name
    df = pd.merge(tmp_tp, rotation, left_on='sub_time', right_on='OUT_TIME_REAL')
    df = pd.merge(df, home_game)
    df['PLAYER_NAME'] = df['PLAYER_FIRST'] + " " + df['PLAYER_LAST']
    df['PLAYER_NAME'] = df['PLAYER_NAME'].apply(lambda x: x.replace(".","")) # clean up names

    # create starter column and sort
    df.loc[df['IN_TIME_REAL'] == 0, 'STARTER'] = True
    df.loc[df['IN_TIME_REAL'] != 0, 'STARTER'] = False
    df = df.sort_values(['IN_TIME_REAL', 'OUT_TIME_REAL'])

    # watch out for whole dataset
    df['obs_id'] = np.arange(1, df.shape[0] + 1)

    # seperate starters
    starter = df[df['STARTER']][['GAME_ID', 'PERSON_ID', 'TEAM_ID','PLAYER_NAME', 'TEAM_NAME', 'is_home_game']].sort_values(['TEAM_NAME', 'is_home_game'])

    # get starting lineup
    starter['PLAYER_ROLE'] = ['AWAY1', 'AWAY2', 'AWAY3', 'AWAY4', 'AWAY5', 'HOME1', 'HOME2', 'HOME3', 'HOME4', 'HOME5']
    starting_lineup = pd.pivot_table(starter, index=['GAME_ID'], columns=['PLAYER_ROLE'], values=['PLAYER_NAME'], aggfunc=np.sum).reset_index()

    # prepare subs
    tmp_subs = pd.melt(df[['IN_TIME_REAL', 'OUT_TIME_REAL', 'obs_id']], id_vars='obs_id')

    # get unique substitutions and create a map
    unique_times = tmp_subs['value'].unique()
    tmp = pd.Series(unique_times).factorize()
    df_dict_subs = pd.DataFrame({'sub_event':tmp[0], 'sub_time':tmp[1]})

    # melt only the two time vars
    id_cols = df.loc[:, (df.columns != 'IN_TIME_REAL') & (df.columns != 'OUT_TIME_REAL')].columns.to_list()
    df2 = pd.melt(df, id_vars=id_cols, value_vars=['IN_TIME_REAL', 'OUT_TIME_REAL'])

    # merge map back to dataframe
    df = pd.merge(df2, df_dict_subs, left_on='value', right_on='sub_time')
    df = df.drop(['sub_time_y'], axis=1)

    # link in sub and out sub
    tmp_s = df[['PERSON_ID', 'PLAYER_NAME', 'variable', 'sub_event']]
    tmp_s = pd.pivot(tmp_s, columns="variable", values="sub_event")
    tmp_s.columns = ['in_sub_event', 'out_sub_event']

    # concatenate back to dataframe
    df = pd.concat([df, tmp_s], axis=1)

    # subset subs dataset from df
    subs = df[['PERSON_ID', 'PLAYER_NAME', 'in_sub_event', 'out_sub_event']]

    # move from data, starters, starting_lineup and subs to dummy matrix with home away weights

    # get all the sub events per game
    sub_events = subs['out_sub_event'].unique()
    sub_events = sub_events[~np.isnan(sub_events)]

    # get last sub event
    last_event = np.max(sub_events)

    # transform the starting lineup to np array
    sl = np.asarray(starting_lineup.values.tolist()[0][1:])

    # init list
    list_lineups = list()

    # start with starting lineup
    list_lineups.append(sl)
    current_lineup = sl

    # loop over substituions
    for s, sub in enumerate(sub_events):

        if sub < last_event:
            # get subs in and subs out
            sub_out = subs[subs['out_sub_event'] == sub]['PLAYER_NAME'].to_numpy()
            sub_in = subs[subs['in_sub_event'] == sub]['PLAYER_NAME'].to_numpy()

            # substitute players
            current_lineup[np.isin(current_lineup, sub_out)] = sub_in

        list_lineups.append(current_lineup.copy())

    # placeholders
    colnames = ["P1", "P2", "P3", "P4", "P5", "P6", "P7", "P8", "P9", "P10"]

    # adapt shape and create on_court dataframe
    arr_on_court = np.concatenate(list_lineups, axis=0).reshape(-1, 10)
    df_on_court = pd.DataFrame(arr_on_court, columns=colnames)
    df_on_court['stint'] = np.insert(sub_events, 0, 0, axis=0)
    df_on_court['GAME_ID'] = game_id

    # melt it
    df_on_court_long = pd.melt(df_on_court, id_vars=['GAME_ID', 'stint'])

    # seperate players who played in this game and get home game flag
    player_team = df[['PLAYER_NAME', 'TEAM_NAME', 'is_home_game']].drop_duplicates()

    # add to dataframe
    tmp = pd.merge(df_on_court_long, player_team, left_on='value', right_on='PLAYER_NAME', how='left')
    tmp = tmp[['PLAYER_NAME', 'is_home_game']].drop_duplicates()

    # transform to dummy matrix
    tmp_dummies = pd.get_dummies(df_on_court_long, columns=['value'], prefix='', prefix_sep='')
    tmp_dummies = tmp_dummies.drop(['variable'], axis=1)

    # groupby stint and sum up
    test = tmp_dummies.iloc[:, 1:].groupby('stint').sum().reset_index()

    # extract player names and store their index
    tmp_player = pd.DataFrame({'PLAYER_NAME':test.columns.tolist()})
    tmp_player = tmp_player.reset_index()
    tmp_player.columns = ['index_pos', 'PLAYER_NAME']

    # merge player and player team to create home away vector
    tmp_vec = pd.merge(tmp_player, player_team, how='left')
    tmp_vec['vector'] = np.where(tmp_vec['is_home_game'], 1, -1)
    tmp_vec = tmp_vec[tmp_vec['PLAYER_NAME'] != 'stint']

    # extract weight vector and dummy matrix
    vec_weight = tmp_vec['vector'].values
    mat_dummy = test.iloc[:, 1:].values

    # insert back into the dataframe
    test.iloc[:, 1:] = mat_dummy * vec_weight

    # add game id
    test['GAME_ID'] = game_id
    
    return test
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from src_dsp.rapm_toolbox import *

# load schedule and all rotation data
schedule_all = pd.read_csv(data_path + "schedule.csv", dtype={'GAME_ID':str})
rotation_all = pd.read_csv(data_path + "gamerotation_16_20.csv", dtype={'GAME_ID':str})

# get the game ids in the rotation data
game_ids = rotation_all['GAME_ID'].unique()

# storage
list_dummys = list()

# loop over all game ids
for game in tqdm(game_ids):
    
    # subset for current game
    rotation = rotation_all[rotation_all['GAME_ID'] == game]
    schedule = schedule[schedule['GAME_ID'] == game]
    
    # call generate dummy function
    tmp = generate_dummy(rotation=rotation, schedule=schedule, game_id=game)

    # store in list
    list_dummys.append(tmp)

# concatenate list and store
df = pd.concat(list_dummys)
df.to_csv(data_path + "dummy_matrix_rotation.csv")
```
