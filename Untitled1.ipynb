```python
import pandas as pd
import numpy as np
```

```python
tmp = pd.read_csv("player_salaries_18_19.csv", sep=";", header=None)
tmp2 = pd.read_csv("player_salaries_17_18.csv", sep=";", header=None)
tmp3 = pd.read_csv("player_salaries_16_17.csv", sep=";", header=None)
data = pd.read_csv("./data/data_assets/players_salaries_hist.csv")
```

```python
tmp['SEASON_ID'] = "2018/19"
tmp2['SEASON_ID'] = "2017/18"
tmp3['SEASON_ID'] = "2016/17"
```

```python
df = pd.concat([tmp, tmp2, tmp3])
```

```python
df.columns = ['player_names', 'SALARY', 'DROP', 'SEASON_ID']
```

```python
df = df.drop('DROP', axis=1)
```

```python
df2 = df.pivot('player_names', 'SEASON_ID', 'SALARY').reset_index()
```
